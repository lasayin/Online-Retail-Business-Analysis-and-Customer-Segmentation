#### Hypothesis Testing

**Null Hypothesis H<sub>0</sub>:** the observed difference of monthly revenue between September and low-revenue months, i.e., the other nine months expect September, October and November, is due to chance.

**Alternative Hypothesis H<sub>1</sub>:** the observed difference of monthly revenue between September and low-revenue months is not due to chance, in other words, due to pre-holiday purchase.

We're also going to pick a **significance level** of 0.05. 

##### September Revenue

```python
# Drop Quantity column and rows of Octomber and November
month_rev = month_qua_rev.drop(index=[10,11,22,23], columns='Quantity').reset_index(drop=True)
```

```python
# Build new column to label if September or not
month_rev['Sep'] = 'N'
index = month_rev[(month_rev.InvoiceDate == '2010-09-30') | (month_rev.InvoiceDate == '2011-09-30')].index
month_rev.loc[index, 'Sep'] = 'Y'
month_rev
```

```python
# Calclatet the difference of average revenue between the low-revenue month and September
sep_revenue = month_rev[month_rev.Sep=='Y']['Revenue']
low_month_revenue = month_rev[month_rev.Sep=='N']['Revenue']
obs_difference = sep_revenue.mean() - low_month_revenue.mean()
obs_difference
```

```python
# Monthly revenue distribution
plt.figure(figsize=(8,5))
sep_revenue.hist(label='September Revenue', bins=10)
low_month_revenue.hist(label='Low-Month Revenue', bins=10)
plt.axvline(sep_revenue.mean(), color='red', label='Average September Revenue')
plt.axvline(low_month_revenue.mean(), color='green', label='Average Low-month Revenue')
plt.legend()
plt.xlabel('Monthly Revenue')
plt.ylabel('Frequency')
_ = plt.title('September Revenue and Low-month Revenue Distribution', size=20)
```

We will use permutation test, which has no assumption of the population distribution.

```python
# Build a list to store the difference of revenue between the low-revenue month and September
difference = []
for _ in range(10000):
    month_rev['Permutational'] = np.random.permutation(month_rev.Revenue)
    difference.append(np.mean(month_rev[month_rev.Sep=='Y'].Permutational) - 
                      np.mean(month_rev[month_rev.Sep=='N'].Permutational))
```

```python
# Plot the difference distribution
plt.axvline(obs_difference, color='red', label='Observed Difference')
plt.legend()
plt.title('Distribution of the Revenue Difference under Null Hypothesis', fontsize=15)
_ = sns.histplot(difference, kde=True, bins=30)
```

```python
p_value = 1 - stats.percentileofscore(difference, obs_difference) * 0.01
print('P value of the observed difference is {:.4f}'.format(p_value))
```

## Time Cohort Analysis

Time cohorts are customers who signed up for a product or service during a particular time frame. Analyzing these cohorts shows the customersâ€™ behavior depending on the time they started using the companyâ€™s products or services. There are other types of cohort , such as behavior cohort and size cohort. For this object, we will focus on time cohort analysis.

Now, we will remove all the transactions without Customer ID, which is labeled as 0.

```python
# Keep transactions with Customer ID only
retail = retail[retail.CustomerID!=0]

(retail.CustomerID==0).sum()
```

### Acquisition Month Cohort

In this project, we group customers into cohorts based on the month of their first purchase(within the timeframe of dataset).

```python
# Find out the date when the customer purchased for the first time
cohort_date = retail.groupby('CustomerID')['InvoiceDate'].transform('min')
# Define cohort based on the month
retail['CohortMonth'] = cohort_date.apply(lambda x: dt(x.year, x.month, 1))
retail['InvoiceMonth'] = retail.InvoiceDate.apply(lambda x: dt(x.year, x.month, 1))
retail
```

```python
retail
```

We will calculate the number of months since the acquisition for every Customer ID

```python
def get_date(df, column):
    # Extract year, month, day from a date
    year = df[column].dt.year
    month = df[column].dt.month
    day = df[column].dt.day
    return year, month, day
```

```python
# Extract the year, month, day of invoice date
invoice_y, invoice_m, invoice_d = get_date(retail, 'InvoiceDate')
# Extract the year, month, day of cohort month
cohort_y, cohort_m, cohort_d = get_date(retail, 'CohortMonth')
# Define cohort index with the number of months since the customer's first purchase
retail['CohortIndex'] = (invoice_y - cohort_y) * 12 + (invoice_m - cohort_m) + 1
retail.head()
```

Now, we can calculate monthly active customers from each cohort

### Monthly Active Customers by Cohort

```python
# Calculate the monthly active customers from each cohort
monthly_cohort = retail.groupby(['CohortMonth', 'CohortIndex']).nunique().CustomerID.to_frame()
monthly_cohort = pd.pivot_table(monthly_cohort, index='CohortMonth', columns='CohortIndex')
monthly_cohort
```

Here, 1 of Cohort Index represents the month, in which the cohort purchase for the first time; 2 represents the following month, and so on.

### Customer Retention

The retention measures how many customers from each of the cohort have returned in the subsequent months. We can calculate the retention rate, i.e., the ratio of how many of these customers came back in the subsequent months.

```python
# Calculate the size of each cohort
cohort_sizes = monthly_cohort.iloc[:,0]
# Calculate the retention rate (percentage)
retention = monthly_cohort.divide(cohort_sizes, axis=0).round(3)
retention
```

```python
# Use the Cohort Index as columns name for heatmap
retention.columns = [a[1] for a in retention.columns.to_flat_index()]

plt.figure(figsize=(16, 14))
plt.title('Monthly Customer Retention Rate', fontsize=20)
_ = sns.heatmap(data=retention, annot=True, fmt='.0%', 
                vmin=0, vmax=0.5, cmap='YlGn', yticklabels=retention.index.strftime('%Y-%m-%d'))
```

The statistics of the first several months of the dataset, especially December 2009, can not present the real world because of the absence of the previous months. According to the heatmap, the retention rates are relatively higher in the pre-holiday season than the other periods.

### Average Monthly Revenue for Each Cohort

```python
revenue_cohort = retail.groupby(['CohortMonth', 'CohortIndex'])['TotalPrice'].sum().reset_index()
revenue_cohort = revenue_cohort.pivot(index='CohortMonth', columns='CohortIndex').divide(1000).round()
revenue_cohort
```

```python
revenue_cohort.columns = [a[1] for a in revenue_cohort.columns.to_flat_index()]

plt.figure(figsize=(16,14))
plt.title('Monthly Revenue (kâ‚¤) Generated by Cohort', fontsize=20)
_ = sns.heatmap(revenue_cohort, cmap='YlGn', annot=True, fmt='.3g',
                vmin=0, vmax=70, yticklabels=revenue_cohort.index.strftime('%Y-%m-%d'))
```

Revenue heatmap has a similar pattern with retention rate heatmap. The revenue from each cohort is higher in pre-holidays season and, of course,  the acquisition month. In the same way,  we need to ignore the first few months since the revenue of these months were generated by all the previous customers which are not included in our dataset.

## Customer Lifetime Value(CLV)

The customer lifetime value is a measurement of how much a company expects to earn from an average customer in a lifetime. It can be historical, and we also can predict the customer lifetime value. Here, we will calculate historical CLV only by summing up each customer's revenue instead of profit since we have no cost data.

### Basic CLV

#### Calculating by Month

```python
# Calculate monthly spend per customer
monthly_revenue = retail.groupby(['CustomerID', 'InvoiceMonth'])['TotalPrice'].sum()
monthly_revenue
```

```python
# Calculate average monthly spend 
monthly_revenue_average = monthly_revenue.mean()
# Set lifespan as 12
lifespan_month = 12
# Calculate CLV
clv_m_basic = round(monthly_revenue_average * lifespan_month, 1)

clv_m_basic
```

### Traditional CLV

The traditional CLV is one of the most popular descriptive customer lifetime value technique. It incorporates retention and churn rates, which is defined as 1 minus retention. The retention to churn ratio can act as a proxy to expected length of the customer lifespan with the company. So here, we don't need to define customer lifespan as we did when calculating basic CLV. 

The traditional CLV assumes that the churn is final. When we calculate traditional CLV based on average monthly revenue, we assume that the customers who don't come back the next month, are not coming back in the later periods. But it's not always the case in the real world, especially in non-contract business.

```python
# The retention rate we have calculated above
retention
```

```python
# Calculate average monthly retention rate
# Exclude the first month, of which the retention rate is 100%
retention_average = retention.iloc[:,1:].mean().mean()
# Calculate traditional CLV
clv_month_tradition = monthly_revenue_average * retention_average / (1 - retention_average)

clv_month_tradition
```

It's much lower than the basic CLV since here we assume that the customers, who don't buy products in the next month, will not come back anymore. 

## Recency, Frequency, MonetaryValue and Tenure

### RFM

Recency, Frequency, MonetaryValue (RFM) are customer behavior metrics to measure customer's behavior :
* Recency: measures how recent was each customer's last purchase
* Frequency: measures how many purchases the customer has done in the past period of time
* Monetary Value - measures how much has the customer spent in the past period of time

Here, we use the whole dataset which includes all the entries in two years, i.e., December 2009 through November 2011. We will take the last  date plus 1, i.e., 12/01/2011, as current date to calculate recency.

```python
# Define the sanpshot date, which is the last date of the dataset plus 1
snapshot_date = dt(2011,12,1)
retail_rfm = retail.groupby(['CustomerID']).agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'Invoice': 'count',
    'TotalPrice': 'sum'})
# Rename columns with RFM
retail_rfm.rename(columns={'InvoiceDate': 'Recency',
                         'Invoice': 'Frequency',
                         'TotalPrice': 'MonetaryValue'}, inplace=True)
```

### Tenure

Tenure defines how long the customer has been with the company since their first transaction. We can get tenure of a customer by calculating the days between the date of first transaction and 12/01/2011, a snapshot date of current date.

```python
# The date of the first transaction
retail['CohortDate'] = retail.groupby('CustomerID')['InvoiceDate'].transform('min')
# Calculate tenture
retail['Tenure'] = (dt(2011,12,1)-retail.CohortDate).apply(lambda x: x.days)
# Build datafram to store RFM plus Tenure by customer
retail_rfmt = retail_rfm.assign()
# Add Tenure column
retail_rfmt['Tenure'] = retail.groupby(['CustomerID'])['Tenure'].mean()
```

### Build Dummy A/B Testing Data with Poisson Distribution

```python
# Check data size 
size = high_all_pur.shape[0]

# Build Poisson distribution with the original dataset mean and size
s = np.random.poisson(high_all_pur.NumPurchase.mean(), size)
# Plot Poisson distribution
_ = plt.hist(s, bins=30)
```

```python
# Shuffle the dataset
shuffled = shuffle(high_all_pur)
# Randomly take half of the customers as control group
con = shuffled.sample(2656)
# Assign the number of purchase with Poisson distribution with the origial mean value
con['NumPurchase'] = np.random.poisson(high_all_pur.NumPurchase.mean(), 2656)
con['Group'] = 'C'

# Take the other half as treatment group
tre = shuffled.drop(index=con.index)
# Assign the number of purchase with Poisson distribution with the origial mean value plus 0.2
tre['NumPurchase'] = np.random.poisson(high_all_pur.NumPurchase.mean() + 0.2, 2657)
tre['Group'] = 'T'

# Concat two groups
test = pd.concat([con, tre]).reset_index(drop=True)
test
```

```python
# Check the dummy number of purchase distribution by group
print(test.groupby('Group').describe().NumPurchase)
_ = sns.histplot(data=test, x='NumPurchase', hue='Group', multiple='dodge', bins=20)
```

### Hypotheses  Testing

**Null Hypotheses ($H_0$):** There is no significant difference between the control group and the treatment group, i.e., the lift of the average number of purchase is random.

**Alternative Hypotheses($H_a$):** There is a significant improvement of number of purchase in treatment group compared with control group.

We can accept or reject Null Hypotheses based on P value.

```python
# Calculate p value
p = stats.ttest_ind(tre.NumPurchase, con.NumPurchase, alternative='greater')[1]
p
```

### Confidence Interval

The pooled variance of the two groups can be estimated by:
\begin{equation}
 Ïƒ_{\bar{x_1} - \bar{x_0}} ^2 =  Ïƒ_{\bar{x_1}}^2 + Ïƒ_{\bar{x_0}}^2 = {\frac{Ïƒ^2_1}{n_1}} + {\frac{Ïƒ^2_0}{n_0}} â‰ˆ {\frac{s^2_1}{n_1}} + {\frac{s^2_0}{n_0}}
\end{equation}

The one-sided interval of *t*-test can be given by:

\begin{equation}
\mu \geq \mu_{\bar{x_1}-\bar{x_0}} - t_{1-\alpha,n-1} Ã— Ïƒ_{\bar{x_1} - \bar{x_0}} = \bar{x_1} - \bar{x_0} - t_{1-\alpha,n-1} Ã— Ïƒ_{\bar{x_1} - \bar{x_0}} .
\end{equation}

Here, $Ïƒ_{\bar{x_1} - \bar{x_0}} ^2$ is the variance of the distribution of the difference of sample mean; $ Ïƒ_{\bar{x_1}}^2 $ and $ Ïƒ_{\bar{x_0}}^2 $ are the variance of sampling distribution of sample $x_1$ and the variance of sampling distribution of sample $x_0$. $\sigma_1^2$ and $\sigma_0^2$ are the variance of population 1 and population 0, which can be estimated by the variance of sample 1 and sample 0; $\alpha$ is significant level, $n-1$ is degree of freedom, $t$ is critical $t$ value, which can be looked up through t table.

```python
# For one-tailed t distribution, critical t value is 1.645 based on 0.05 significant level.
t = 1.645
# Calculate the difference of the mean values between two groups
delta = tre.NumPurchase.mean() - con.NumPurchase.mean()
# Calculate the size of two groups
tre_size=tre.shape[0]
con_size=con.shape[0]
# Calculate the standard deviation of two groups
tre_std = np.std(tre.NumPurchase,ddof=1)
con_std = np.std(con.NumPurchase,ddof=1)
# Calculate the 95% 
ci = delta - 1.645 * np.sqrt(tre_std ** 2 / tre_size + con_std ** 2 / con_size)
print('There is a 95% chance that the increase in the average number of purchase is greater than {:.3f}.'.format(ci))
```

```python
from scipy import stats
diff_mean = delta
diff_std = np.sqrt(tre_std ** 2 / tre_size + con_std ** 2 / con_size)
diff_line = np.linspace(-3*diff_std+diff_mean, 3*diff_std+diff_mean, 100)
```

```python
plt.plot(diff_line, stats.norm.pdf(diff_line, diff_mean, diff_std), lw=2)
plt.ylim(0,14)
x_shade = np.arange(ci,3*diff_std+diff_mean, 0.005)
y_shade = stats.norm.pdf(x_shade, diff_mean, diff_std)
plt.axvline(ci, color='g', lw=2, label=f'{ci.round(3)}')
plt.fill_between(x_shade,y_shade,color='b')
plt.legend()
_ = plt.title('95% Confidence Interval', fontsize=15)
```
