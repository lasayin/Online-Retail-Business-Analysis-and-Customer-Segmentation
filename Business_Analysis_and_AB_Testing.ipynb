### Hypothesis Testing

We're also going to pick a **significance level** of 0.05.

#### Permutation Test

The data size was super small. There were 2 data points representing September (2010/09 and 2011/09) and 16 data points representing other 8 months (2010/01 through 2010/08 and 2011/01 through 2011/08). Permutation Test does not require you to make any assumptions about the sampling distribution (e.g., it doesn't require the samples to be normally distributed), so it was a good choice for this case. 

There were a total of (2+16)! permutations and I randomly selected 10k of them.

```python
def permutation_test(df, month, col):
    # Implement permutation test
    # Display result
    # Calculate p-value
    
    # Calculate observed difference between the target month and the other months
    obs_diff = df[df[month]=='Y'][col].mean() - df[df[month]=='N'][col].mean()

    # Create a list to store the permutational difference of revenue
    difference = []
    
    for _ in range(10000):
        # Randomly draw values from all possible permutations
        df['Permutational'] = np.random.permutation(df[col])
        # Calculate the difference of permutational value between the target month and the other months
        # Append all results to the difference list
        difference.append(np.mean(df[df[month]=='Y'].Permutational) - 
                          np.mean(df[df[month]=='N'].Permutational))
        
    # Plot the difference distribution
    plt.figure(figsize=(10,6))
    plt.axvline(obs_diff, color='red', label='Observed Difference')
    plt.legend()
    plt.title(f'{month} {col} Hypothesis Testing', fontsize=15)
    _ = sns.histplot(difference, kde=True, bins=30)
    
    p_value = 1 - stats.percentileofscore(difference, obs_diff) * 0.01
    print('P value for {} {} Hypothesis Testing is {:.4f}.'.format(month, col, p_value))
    
    return p_value
```

```python
p_sep_rev = permutation_test(rev_9m, 'September', 'Revenue')
```

The calculated p value is less than the predefined significant level, i.e., 0,05, so we can reject the null hypothesis.

## A/B Testing

### Build Control and Treatment Group

```python
# Create Invoice Month column 
retail['InvoiceMonth'] = retail.InvoiceDate.apply(lambda x: dt(x.year, x.month, 1)).apply(lambda x: x.strftime('%Y-%m-%d'))
retail.sample(5)
```

```python
# Extract 2011 Sep, Oct, and Nov as peak season
peak = retail[retail.InvoiceMonth.apply(lambda x: x in ['2011-09-01','2011-10-01', '2011-11-01'])]
peak.head()
```

```python
# Calculate the number of purchase by customer id
purchase_peak_active = peak.groupby('CustomerID').nunique().Invoice.reset_index()
purchase_peak_active.head()
```

```python
# Remove the non-sense customer id, which was imputed with 0 in data wrangling step, and rename
purchase_peak_active = purchase_peak_active[purchase_peak_active.CustomerID != 0].rename(columns = {'Invoice':'Purchase'})
purchase_peak_active.head()
```

```python
# Calculate the number of purchase by customer id in the original dataset
purchase_total = retail.groupby('CustomerID').nunique().Invoice.reset_index()
# Remove the non-sense customer id and rename
purchase_total = purchase_total[purchase_total.CustomerID != 0].rename(columns = {'Invoice':'Purchase'})
purchase_total.head()
```

```python
# Build a list to store the customer who did not purchase anything in the 2011 Sep, Oct, and Nov
inactive_cus = [customer for customer in purchase_total.CustomerID.unique() if customer not in purchase_peak_active.CustomerID.unique()]

# Built a dataframe of inactive customer with 0 purchase
inactive_cus = pd.DataFrame({'CustomerID':inactive_cus, 'Purchase':0})
inactive_cus
```

```python
# Concatenate the purchase of active customer and inactive customer(0 purchase) in peak season
purchase_peak = pd.concat([inactive_cus, purchase_peak_active], ignore_index=True)

# Check the number of purchase distribution in the peak season, including active and inactive customers
print(purchase_peak.Purchase.describe())
_ = purchase_peak.Purchase.hist(bins=100)
```

The distribution is extremely right-skewed. We will remove the top 1% to decrease the variance.

```python
# Remove the top 1% 
top_1 = purchase_peak.Purchase.quantile(0.99)
purchase_peak = purchase_peak[purchase_peak.Purchase < top_1]
# Check the distribution of number of purchase without the top 1% purchase
print(purchase_peak.Purchase.describe())
_ = purchase_peak.Purchase.hist(bins=30)
```

The purchase distribution is similar to exponential distribution.

```python
# Check data size 
size = purchase_peak.shape[0]

# Build exponential distribution with the original dataset mean and size
expo = np.random.exponential(purchase_peak.Purchase.mean(),size)
# Plot Poisson distribution
_ = plt.hist(expo,bins=30)
```

We can simulate the purchase distribution with exponential distribution. The main problem is number of purchase is discrete, while exponential distribution is continues. Another way is to use Poisson distribution, a type of discrete distribution.

```python
# Build poisson distribution with the original dataset mean and size
poisson = np.random.poisson(purchase_peak.Purchase.mean(),size)
# Plot Poisson distribution
_ = plt.hist(poisson,bins=30)
```

In this A/B Testing, I assumed that, in the treatment group, a 10% lift of the average number of purchases was observed following a marketing campaign. The control group and treatment group were randomly extracted from the shuffled original dataset, with the same size.

```python
# Check the data size
print(size)
```

```python
# Shuffle the dataset
shuffled = shuffle(purchase_peak)
# Randomly take half of the customers as control group
control = shuffled.sample(2624)
# Assign the number of purchase with Poisson distribution with the origial mean value
control['Purchase'] = np.random.poisson(purchase_peak.Purchase.mean(), 2624)
control['Group'] = 'C'

# Take the other half as treatment group
treat = shuffled.drop(index=control.index)
# Assign the number of purchase with Poisson distribution with 10% lift
treat['Purchase'] = np.random.poisson(purchase_peak.Purchase.mean() * 1.1, 2624)
treat['Group'] = 'T'

# Concat two groups
test = pd.concat([control, treat]).reset_index(drop=True)
test
```

```python
# Check the dummy number of purchase distribution by group
print(test.groupby('Group').describe().Purchase)
_ = sns.histplot(data=test, x='Purchase', hue='Group', multiple='dodge', bins=20)
```

### Hypotheses  Testing

**Null Hypotheses ($H_0$):** There is no significant difference between the control group and the treatment group, i.e., the lift of the average number of purchase is random.

**Alternative Hypotheses($H_a$):** There is a significant improvement of number of purchase in treatment group compared with control group.

We can accept or reject Null Hypotheses based on P value.

```python
# Calculate p value
p = stats.ttest_ind(treat.Purchase, control.Purchase, alternative='greater')[1]
p
```

The p value is less than 0.05, so we can reject null hypothesis. There is a significant lift of the number of purchase.

Now, let's calculate confidence interval. We can consider this case as a one-tail t test.

### Confidence Interval

The pooled variance of the two groups can be estimated by:
\begin{equation}
 σ_{\bar{x_1} - \bar{x_0}} ^2 =  σ_{\bar{x_1}}^2 + σ_{\bar{x_0}}^2 = {\frac{σ^2_1}{n_1}} + {\frac{σ^2_0}{n_0}} ≈ {\frac{s^2_1}{n_1}} + {\frac{s^2_0}{n_0}}
\end{equation}

The one-sided interval of *t*-test can be given by:

\begin{equation}
\mu \geq \mu_{\bar{x_1}-\bar{x_0}} - t_{1-\alpha,n-1} × σ_{\bar{x_1} - \bar{x_0}} = \bar{x_1} - \bar{x_0} - t_{1-\alpha,n-1} × σ_{\bar{x_1} - \bar{x_0}} .
\end{equation}

Here, $σ_{\bar{x_1} - \bar{x_0}} ^2$ is the variance of the distribution of the difference of sample mean; $ σ_{\bar{x_1}}^2 $ and $ σ_{\bar{x_0}}^2 $ are the variance of sampling distribution of sample $x_1$ and the variance of sampling distribution of sample $x_0$. $\sigma_1^2$ and $\sigma_0^2$ are the variance of population 1 and population 0, which can be estimated by the variance of sample 1 and sample 0; $\alpha$ is significant level, $n-1$ is degree of freedom, $t$ is critical $t$ value, which can be looked up through t table.

```python
# Calculate critical t value, based on 0.05 significant level
t = stats.t.ppf(0.95, size-1)
# Calculate the difference of the mean values between two groups
diff_mean = treat.Purchase.mean() - control.Purchase.mean()
# Calculate the size of two groups
treat_size=treat.shape[0]
control_size=control.shape[0]
# Calculate the standard deviation of two groups
treat_std = np.std(treat.Purchase,ddof=1)
control_std = np.std(control.Purchase,ddof=1)
# Calculate the 95% ci
ci = diff_mean - t * np.sqrt(treat_std ** 2 / treat_size + control_std ** 2 / control_size)
print('There is a 95% chance that the increase in the average number of purchase is greater than {:.3f}.'.format(ci))
```

```python
# dispaly the result
from scipy import stats
diff_std = np.sqrt(treat_std ** 2 / treat_size + control_std ** 2 / control_size)
diff_line = np.linspace(-3*diff_std+diff_mean, 3*diff_std+diff_mean, 100)

plt.plot(diff_line, stats.norm.pdf(diff_line, diff_mean, diff_std), lw=2)
plt.ylim(0,16)
x_shade = np.arange(ci,3*diff_std+diff_mean, 0.005)
y_shade = stats.norm.pdf(x_shade, diff_mean, diff_std)
plt.axvline(ci, color='g', lw=2, label=f'{ci.round(3)}')
plt.fill_between(x_shade,y_shade,color='b')
plt.legend()
_ = plt.title('95% Confidence Interval', fontsize=15)
```

## Time Cohort Analysis

Time cohorts are customers who signed up for a product or service during a particular time frame. Analyzing these cohorts shows the customers’ behavior depending on the time they started using the company’s products or services. There are other types of cohort , such as behavior cohort and size cohort. For this object, we will focus on time cohort analysis.

### Month Cohort

In this project, we group customers into cohorts based on the month of their first purchase(within the timeframe of dataset). We have created Invoice Month column, we will create cohort month column based on invoice date

```python
# Find out the date when the customer purchased for the first time
cohort_date = retail.groupby('CustomerID')['InvoiceDate'].transform('min')
# Define cohort based on the month
retail['CohortMonth'] = cohort_date.apply(lambda x: dt(x.year, x.month, 1))
retail.head()
```

We will calculate the number of months since the acquisition for every Customer ID

```python
def get_date(df, column):
    # Extract year, month, day from a date
    
    year = df[column].dt.year
    month = df[column].dt.month
    day = df[column].dt.day
    
    return year, month, day
```

```python
# Extract the year, month, day of invoice date
invoice_y, invoice_m, invoice_d = get_date(retail, 'InvoiceDate')
# Extract the year, month, day of cohort month
cohort_y, cohort_m, cohort_d = get_date(retail, 'CohortMonth')
# Define cohort index with the number of months since the customer's first purchase
retail['CohortIndex'] = (invoice_y - cohort_y) * 12 + (invoice_m - cohort_m) + 1
retail.head()
```

Now, we can calculate monthly active customers from each cohort

### Monthly Active Customers by Cohort

```python
# Calculate the monthly active customers from each cohort
monthly_cohort = retail.groupby(['CohortMonth', 'CohortIndex']).nunique().CustomerID.to_frame()
monthly_cohort = pd.pivot_table(monthly_cohort, index='CohortMonth', columns='CohortIndex')
monthly_cohort
```

Here, 1 of Cohort Index represents the month, in which the cohort purchase for the first time; 2 represents the following month, and so on.

### Customer Retention

The retention measures how many customers from each of the cohort have returned in the subsequent months. We can calculate the retention rate, i.e., the ratio of how many of these customers came back in the subsequent months.

```python
# Calculate the size of each cohort
cohort_sizes = monthly_cohort.iloc[:,0]
# Calculate the retention rate (percentage)
retention = monthly_cohort.divide(cohort_sizes, axis=0).round(3)
retention
```

```python
# Use the Cohort Index as columns name for heatmap
retention.columns = [a[1] for a in retention.columns.to_flat_index()]

plt.figure(figsize=(16, 14))
plt.title('Monthly Customer Retention Rate', fontsize=20)
_ = sns.heatmap(data=retention, annot=True, fmt='.0%', 
                vmin=0, vmax=0.5, cmap='YlGn', yticklabels=retention.index.strftime('%Y-%m-%d'))
```

The statistics of the first several months of the dataset, especially December 2009, can not present the real world because of the absence of the previous months. According to the heatmap, the retention rates are relatively higher in the pre-holiday season than the other periods.

### Average Monthly Revenue for Each Cohort

## Recency, Frequency, MonetaryValue and Tenure

### RFM

Recency, Frequency, MonetaryValue (RFM) are customer behavior metrics to measure customer's behavior :
* Recency: measures how recent was each customer's last purchase
* Frequency: measures how many purchases the customer has done in the past period of time
* Monetary Value - measures how much has the customer spent in the past period of time

Here, we use the whole dataset which includes all the entries in two years, i.e., December 2009 through November 2011. We will take the last  date plus 1, i.e., 12/01/2011, as current date to calculate recency.

```python
# Define the sanpshot date, which is the last date of the dataset plus 1
snapshot_date = dt(2011,12,1)
retail_rfm = retail.groupby(['CustomerID']).agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'Invoice': 'count',
    'Revenue': 'sum'})
# Rename columns with RFM
retail_rfm.rename(columns={'InvoiceDate': 'Recency',
                         'Invoice': 'Frequency',
                         'Revenue': 'MonetaryValue'}, inplace=True)
```

```python
retail_rfm.head()
```

```python
# Check the distribution
print(retail_rfm.describe())
_ = sns.pairplot(retail_rfm, diag_kind='kde')
```

As we can see, the recency, frequency and monetary value are all right-skewed. We will use them to do customer segmentation in the following step and before that, we will un-skew the data.

### Tenure

Tenure defines how long the customer has been with the company since their first transaction. We can get tenure of a customer by calculating the days between the date of first transaction and 12/01/2011, a snapshot date of current date.

```python
# The date of the first transaction
retail['CohortDate'] = retail.groupby('CustomerID')['InvoiceDate'].transform('min')
# Calculate tenture
retail['Tenure'] = (dt(2011,12,1)-retail.CohortDate).apply(lambda x: x.days)
# Build datafram to store RFM plus Tenure by customer
retail_rfmt = retail_rfm.assign()
# Add Tenure column
retail_rfmt['Tenure'] = retail.groupby(['CustomerID'])['Tenure'].mean()
```

Different from RFM, Tenure is left-skewed.

Recency, Frequency, Monetary Value(RFM) and Tenure will be un-skwed in the following preprocessing step and used as metrics for customer segmentation.
